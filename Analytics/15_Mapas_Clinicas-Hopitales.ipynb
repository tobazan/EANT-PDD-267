## <img src="http://icons.iconarchive.com/icons/cornmanthe3rd/plex/256/Other-python-icon.png" width="40"> Ejercicio Hospitales y Cl√≠nicas ü©∫üíâüè•


Generar un mapa de la ciudad de Buenos Aires, donde se muestren con diferentes iconos:
* las cl√≠nicas privadas `[centros-de-salud-privados.csv]`
* los hospitales `[hospitales.csv]`

El mapa debe tener:
* zoom=10, 
* latitud y longitud que sea el promedio de ambas variables del dataset.
* Generar clusters para la visualizaci√≥n.


```
priv = pd.read_csv('./data/centros-de-salud-privados.csv')
pub = pd.read_csv('./data/hospitales_caba.csv', index_col=0)
pub.drop('Unnamed: 0.1', axis=1, inplace=True)
```

```
hp = pub[['NOM_MAP','long','lat']]
hp.columns = ['nombre', 'long', 'lat']
hp['categoria'] = ['publico' for i in range(0, hp.shape[0])]
```

```
cp = priv[['nombre','long','lat']]
cp['categoria'] = ['privado' for i in range(0, cp.shape[0])]
```

```
df = hp.append(cp, ignore_index=True)
df.head()
```

```
df.groupby('categoria')['nombre'].count()
```

```
lat_ini = -34.6219662
long_ini = -58.4493826
zoom = 11

map = folium.Map(location= [ lat_ini , long_ini] , zoom_start= zoom)

for i in range(0, df.shape[0]):
  folium.CircleMarker(location=[df.loc[i,'lat'], df.loc[i,'long']],
                    color='red', radius=20, fill=True, fill_color='red').add_to(map)

  if df.loc[i,'categoria']=='publico':
    folium.Marker([df.loc[i,'lat'], df.loc[i,'long']], popup=df.loc[i,'nombre'],
                icon=folium.Icon(icon='plus-sign', color='green')).add_to(map)
    
  else:
     folium.Marker([df.loc[i,'lat'], df.loc[i,'long']], popup=df.loc[i,'nombre'],
                icon=folium.Icon(icon='plus',color='blue')).add_to(map)

map
```
